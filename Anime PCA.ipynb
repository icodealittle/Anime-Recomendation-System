```python
def rating(x):
    if x == 'PG-13 - Teens 13 or older':
        return 1
    elif x == 'R - 17+ (violence & profanity)':
        return 2
    elif x == 'G - All Ages':
        return 3
    elif x == 'R+ - Mild Nudity':
        return 4
    elif x == 'PG - Children':
        return 5
    elif x == 'None':
        return 0

df_merge['rating'] = df_merge['rating'].apply(rating)
```

```python
first_comp = pca39.components_[0]
first_comps = pd.DataFrame(list(zip(first_comp, new_mer.columns)), columns=['weights', 'features'])
first_comps['abs_weights']=first_comps['weights'].apply(lambda x: np.abs(x))
first_comps.sort_values('abs_weights', ascending=False,inplace=True)

Top_39_features =first_comps
pd.set_option('display.max_rows', Top_39_features.shape[0]+1)
print(Top_39_features)
```
